```
product1 = [10,20,30,40,50] #ยอดขายสินค้าชิ้นที่ 1
product2 = [15,30,10,20,60] #ยอดขายสินค้าชิ้นที่ 2
month = [1,2,3,4,5]
plt.plot(month, product1,"r+-")
plt.plot(month, product2,"g*-")
data = {"size":20,"color":"black"}
plt.xlabel("Month",data)
plt.ylabel("Sales",data)
config = {"size":20,"color":"blue"}
plt.title("TopSale",config)
plt.text(2,32,"point1",size = "20",color = "red")
plt.text(4,15,"point2")
plt.show()
```

```
import matplotlib.pyplot as plt

student = [15,30,45,60,40]
Course = ["php","Flutter","Nodejs","c#","java"]
color = ["red","green","black","purple","cyan"]

exp = [0,1,0,0.2,0] # หยิบกราฟออกมา เลือกมากี่ส่วนก็ได้
plt.pie(student, labels=Course, autopct="%.1f%%", colors=color,shadow = True,explode=exp,startangle=90) #เรียกใช้สี และ ใส่เงา //คำสั้งเรียกใช้ หยิบกราฟ explode=exp//# ปรับมุม 90 องศา startangle=90
plt.show()
```

```
import matplotlib.pyplot as plt

course = ["Python","PHP","java"]
score = [80,50,99]
c = ["red","green","yellow"]

plt.barh(course, score, color=c,edgecolor ="black",linewidth = 3 ) # กำหนดความกว้าง width=3 #เส้นขอบของบา แต่ละแท่ง edgecolor ="black" #ความหนาเส้นขอบlinewidth = 3
plt.xlabel("Score")
plt.ylabel("Course")
plt.title("Score 2021")
plt.show()
```

```
room = ["A","B","c"] # ห้องเรียน
boys = [10,15,5] # นักเรียนชาย
girls = [20,7,15] # นักเรียนหญิง

plt.bar(room,boys,label = "Boys",color = "red")
plt.bar(room,girls,bottom = boys,label = "girls",color = "blue") # เก็บข้อมูลชายไว้ด้านล่าง

plt.legend()

```

```
room = ["A","B","c"] # ห้องเรียน
boys = [10,15,5] # นักเรียนชาย
girls = [20,7,15] # นักเรียนหญิง

plt.stackplot(room,boys,girls,labels=["boys","girls"],colors=["red","blue"])

plt.legend()
```
